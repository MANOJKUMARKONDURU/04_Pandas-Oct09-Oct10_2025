#What is a series
A Pandas Series is like a column in a table.

It is 1-D array holding data of any type

```python
import pandas as pd
```

```python
a = [1, 21, 27]

myvar = pd.Series(a)
print(myvar)
```

#Labels
If nothing else is specified, the values are labeled with their index number.First value has index 0, second values has index 1, etc.

This label can be used to access a specified value.

```python
print(myvar[1])
print(myvar[2])
```

#Create Labels
with the **index** argument, you acn name your own labels.

```python
a = [1, 21, 27]
myVar = pd.Series(a, index = ['x', 'y', 'z'])
print(myVar)
print(myVar["z"])
```

#Key/Value Objects as Series

You can also use a key/vlaue object, like a dictonary, when creating a Series

```python
calories ={
    'day1' : 100,
    'day2' : 200,
    'day3' : 300
}
myVar = pd.Series(calories)
print(myVar)
print(myVar[0])
```

#Data Frames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

**Series is like a column, A DataFrame is the whole table**

```python
data ={
    'calories' : [100, 200, 300],
    'duration' : [50, 40, 45]
}
myVar = pd.DataFrame(data)
print(myVar)
print(myVar['calories'])
```
