#What is a DataFrame?

A Pandas DataFrame is 2-D data structure, like a 2 dimensional array, or a table with rows and columns.

```python
import pandas as pd
```

```python

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

#Locate Row
Pandas use the **loc** attribute to return one or more specified row(s).

**note**:This example returns a Pandas Series

```python
print(df.loc[0])
```

return row 0 and 1:

**Note**: When using**[ ]** result is a Pandas DataFrame

```python
print(df.loc[[0, 1]])
```

#Named Indexes
with the **index** argument, you can name your own indexes.

```python
df = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])

print(df)
```

#Locate Named Indexes

Use the named index in **loc** attribute to return the specied row(s).

```python
print(df.loc['day2'])
```

#Load Files into DataFrame
If your data sets are stored in a file, Pandas can load them into a DataFrame.

```python
df = pd.read_csv('data.csv')

print(df)
```
